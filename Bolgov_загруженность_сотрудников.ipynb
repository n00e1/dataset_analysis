## №1 <a id='section_date1'></a>
По данным из файла "Загрузка_июнь 2022 test" посчитайте месячную загрузку Отделения по должностям, как процент времени приёма от времени на рабочем месте

Структура данных  у нас дерево т е в активе есть отделения в отделениях  есть специализация, в специализациях специальность а в специальности работают люди. Следовательно чтобы посчитать загрузку нам нужно проссумировать столбец **Длительность с учетом перерыва (мин.)**, а также столбец **Фактическая загрузка (в мин.)**. Дальше мы составляем пропорцию<br> 

Пропорция<br> 
**Длительность с учетом перерыва (мин.)** - 100%<br>
**Фактическая загрузка (в мин.)** - x<br>


```python
one = zagryzka_test.groupby(['Asset','Department','Specialization','Position'])[['Duration_including_break_(min.)','Actual_load_(in minutes)']].sum()
one.head(10)
```

```python
one['Monthly_Load'] = one['Actual_load_(in minutes)'] * 100 / one['Duration_including_break_(min.)']
one 
```

## №3 <a id='section_date3'></a>
Добавьте столбец в объединенную таблицу из задания №2 по следующему условию: если загрузка отделения выше или равна 60%, то "+", если ниже, то "-" 

```python
one['plus_minus'] = np.where(one['Monthly_Load'] >= 60, "+", "-")
one 
```

## №4 <a id='section_date4'></a>
Проанализируйте, сколько ставок (кратно 0,25) можно добавить в отделения с загрузкой выше или равной 60%, чтобы сделать её ниже 60%. В новый столбец запишите, сколько ставок можно добавить. (Месячная норма на 1 ставку для всех должностей счиать равной 160 часов)

1 ставка - 160 часов<br> 
0,25 ставка - 40 часов<br>
Так как все измеряется в минутах  то 40 часов - 2400 минут<br> 

Чтобы сделать загруженность ниже 60% нужно увеличить **Длительность с учетом перерыва (мин.)**.<br> 
Составим пропорцию:<br>
x               -               100%<br>
Фактическая загрузка (в мин.) - 59%<br>
Из полученного числа мы должны вычесть прошлую посчитанную сумму по столбцу **Длительность с учетом перерыва (мин.)**

```python
four['x'] = (four['Actual_load_(in minutes)'] * 100 / 59) - four['Duration_including_break_(min.)']
four
```

```python
four['plus_stavka_0.25'] = four['x'] / 2400 # чтобы получить кол-во ставок 0,25 
four
```
